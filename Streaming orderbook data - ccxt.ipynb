```python
#Importing reqd libraries
import asyncio
import os
import sys
import pandas as pd
import ccxt.async_support as ccxt  # noqa: E402


async def main(symbol):
    #Streams data at the rate limit set by the exchange (binance)
    exchange = ccxt.binance({
        'enableRateLimit': True,  # this option enables the built-in rate limiter 
    })
    # Change loop structure to while True for continuous streaming
    for i in range(0,10):
        print('--------------------------------------------------------------')
        print(exchange.iso8601(exchange.milliseconds()), 'fetching', symbol, 'ticker from', exchange.name)
        # exchange.milliseconds ensures timestamp is accurate
        ticker = await exchange.fetch_l2_order_book(symbol)
        print(exchange.iso8601(exchange.milliseconds()), 'fetched', symbol, 'ticker from', exchange.name)
        print(ticker)
        # need to update dataFrame to include timestamp data from 
        orderbook = pd.DataFrame(ticker)
        orderbook['timestamp'] = exchange.iso8601(exchange.milliseconds())
        orderbook.to_csv('orderbook_binance.xlsx',mode='a')
        print(orderbook)
        #currently the orderbook is arranged according to the rate limited by the particular exchange
        # we could sample hourly or minute wise, as required using the timestamps
        #moreover the datetime column is currently None but could be used later for daily aggregates
#error if any, to be ignored, can add exception later on
asyncio.get_event_loop().run_until_complete(main('ETH/BTC'))
```
